```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
gradient = pd.read_csv('gradient_output.csv')
newton = pd.read_csv('newton_output.csv')
perturb = pd.read_csv('perturb_output.csv')

plt.plot(gradient['Episode'], gradient['Reward Mean'], label='Gradient Descent')
plt.fill_between(gradient['Episode'], gradient['Reward Mean'] - 0.5*gradient['Reward Std'], gradient['Reward Mean'] + 0.5*gradient['Reward Std'], alpha=0.2)
plt.plot(newton['Episode'], newton['Reward Mean'], label='Newton\'s Method')
plt.fill_between(newton['Episode'], newton['Reward Mean'] - 0.5*newton['Reward Std'], newton['Reward Mean'] + 0.5*newton['Reward Std'], alpha=0.2)
plt.plot(perturb['Episode'], perturb['Reward Mean'], label='Perturb Descent')
plt.fill_between(perturb['Episode'], perturb['Reward Mean'] - 0.5*perturb['Reward Std'], perturb['Reward Mean'] + 0.5*perturb['Reward Std'], alpha=0.2)
plt.xlabel('Episode')
plt.ylabel('Episodic Reward')
plt.legend()
plt.savefig('output.png')
plt.show()
```
